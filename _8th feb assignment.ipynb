```python
Question 1. What is Abstraction in OOps? Explain with an example?
ANSWER:Abstraction is a fundamental concept in Object-Oriented Programming (OOP) 
that involves simplifying complex reality by modeling classes based
on the essential attributes and behaviors of real-world entities.
It allows us to focus on the relevant characteristics of an object while ignoring unnecessary details. 
Abstraction in OOP provides a way to create abstract data types
that can be used to model and manipulate real-world concepts.
FOR EXAMPLE:
    class Vehicle:
    def __init__(self, name, speed):
        self.name = name
        self.speed = speed

    def move(self):
        print(f"{self.name} is moving at {self.speed} km/h")

# Specific vehicle types
class Car(Vehicle):
    def __init__(self, name, speed):
        super().__init__(name, speed)

class Motorcycle(Vehicle):
    def __init__(self, name, speed):
        super().__init__(name, speed)

class Truck(Vehicle):
    def __init__(self, name, speed):
        super().__init__(name, speed)

# Using the abstraction
car = Car("Sedan", 120)
motorcycle = Motorcycle("Harley", 150)
truck = Truck("Freightliner", 80)

car.move()
motorcycle.move()
truck.move()
```

```python
QUESTION 2: Differentiate between Abstraction and Encapsulation. Explain with an example.
ANSWER:
    #ABSTRACTION:
    Abstraction involves simplifying complex reality by modeling classes based
    on the essential attributes and behaviors of real-world entities.
    It allows US to create abstract data types that represent high-level concepts,
    hiding unnecessary details.
    >It focuses on defining the essential features of an object while hiding the implementation details.
     >It provides a clear and simplified view of an object.
    >It used to create a blueprint for objects.
     >It aims to answer "what" an object does rather than "how" it does it.
        
        for example:
            from abc import ABC, abstractmethod

class Shape(ABC):
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def calculate_area(self):
        pass

class Circle(Shape):
    def __init__(self, name, radius):
        super().__init__(name)
        self.radius = radius

    def calculate_area(self):
        return 3.14 * self.radius ** 2

class Square(Shape):
    def __init__(self, name, side):
        super().__init__(name)
        self.side = side

    def calculate_area(self):
        return self.side ** 2
    
    #ENCAPSULATION:
    Encapsulation involves bundling data (attributes)
    and the methods (functions) that operate on the data into a single unit,
    often called a class. It restricts direct access to some of the
    object's components and promotes the use of methods to access and manipulate the data.
    
    >It focuses on information hiding and data protection.
    > It uses access modifiers (like public, private, protected) to control access to data members.
    >It prevents unintended modification of object state.
     >It provides a level of security and control over data.
        
        for example:
            class Person:
    def __init__(self, name, age):
        self.__name = name  # Private attribute
        self.__age = age    # Private attribute

    def get_name(self):
        return self.__name

    def set_name(self, new_name):
        self.__name = new_name

    def get_age(self):
        return self.__age

    def set_age(self, new_age):
        if new_age > 0:
            self.__age = new_age

person = Person("Alice", 30)
print(person.get_name())  # Access using getter
person.set_age(31)        # Modify using setter
```

```python
Q3. What is abc module in python? Why is it used?
ANSWER: The abc module in Python stands for "Abstract Base Classes."
It provides tools to work with abstract classes and interfaces,
allowing you to define abstract base classes and enforce
the implementation of certain methods in derived classes.
It is a way to introduce a level of structure and hierarchy
into class design, making it easier to ensure that
classes adhere to certain patterns and provide specific functionality.

The abc module is used for the following purposes:
1)Defining Abstract Base Classes:They serve as templates that define a set of methods
that derived classes must implement. The abc module allows you to define abstract 
methods using the @abstractmethod decorator.
2)Enforcing Method Implementation: By using abstract base classes,.
we can enforce that derived classes implement specific methods.
If a subclass fails to implement a required method, attempting to instantiate it will raise an error.
3) Polymorphism and Type Checking: Abstract base classes enable polymorphism,
allowing us to treat different objects that share a common interface in a uniform way.

for example:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

circle = Circle(5)
rectangle = Rectangle(4, 6)

shapes = [circle, rectangle]

for shape in shapes:
    print(f"Area: {shape.area()}")
```

```python
Q4. How can we achieve data abstraction?
ANSWERS: Data abstraction can be achieved in python through the use of classes and objects.
we can achieve data abstraction:
1)Define a Class: Create a class that represents the abstract concept you want to model.
This class should define the essential attributes and methods that characterize the concept.

2) Encapsulate Data: Encapsulate the data attributes within the class.
Use access modifiers to control the visibility and access of these attributes from outside the class.

3) Define Methods: Define methods that provide operations related to the data attributes.
These methods can be used to modify or interact with the data in a controlled manner.

4) Hide Implementation Details: Expose only the necessary methods and attributes to the user
while hiding the complex implementation details. This simplifies the interaction with the object
and reduces the complexity for the user.

5) Provide a Public Interface: Define a clear public interface for the class. 
This interface should consist of methods that represent the
intended use of the class and its objects. The user should interact with
the class through this interface.

for example:
    class BankAccount:
    def __init__(self, account_number, balance):
        self.__account_number = account_number
        self.__balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.__balance}")
        else:
            print("Insufficient funds")

    def get_balance(self):
        return self.__balance

# Create an instance of BankAccount
account = BankAccount("123456789", 1000)

# Interact with the object using the public interface
account.deposit(500)
account.withdraw(200)
print("Current balance:", account.get_balance())
```

```python
Q5. Can we create an instance of an abstract class? Explain your answer.
ANSWER:No, we can not create an instance of an abstract class in Python.
Abstract classes are meant to be subclassed, and they serve as templates
for creating more specific classes. They are designed to be incomplete and
are intended to be extended by concrete subclasses that provide implementations
for the abstract methods defined in the abstract class.

An abstract class is created using the abc modules ABC class as the base class
and using the @abstractmethod decorator to define abstract methods.
An abstract method is a method that has no implementation in
the abstract class and must be implemented by any concrete subclass.

for example:
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

# This would raise an error since Shape is an abstract class
# shape = Shape()

circle = Circle(5)
print("Circle area:", circle.area())
```
